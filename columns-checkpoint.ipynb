```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy.linalg
```

```python
# read file
path = r'C:\Users\HP\OneDrive\Документы\Python\FlowSensor\c1-25.6C-end_p.txt'

x_data = []
y_data = []
z_data = []

data = []
with open(path,"r") as f:
    for line in f:
        data.append([float(x) for x in line.split()])

# forming an array
a = np.array(data)

x_data = a[:, 1]
y_data = a[:, 2]
z_data = a[:, 0]
```

```python
# regular grid covering the domain of the data
data = np.c_[x_data, y_data, z_data]
X,Y = np.meshgrid(np.linspace(min(x_data), max(x_data), 1001), np.linspace(min(y_data), max(y_data), 1001))
XX = X.flatten()
YY = Y.flatten()
```

```python
# order = 'ideal'    # 3: qubic, 4: 4th order, 'ideal': ideal

A_1 = np.c_[data[:,0], data[:,1], np.ones(data.shape[0])]
C_1,_,_,_ = scipy.linalg.lstsq(A_1, data[:,2])    # coefficients
    
Z_1 = np.dot(np.c_[XX, YY, np.ones(XX.shape)], C_1).reshape(X.shape)

A_2 = np.c_[np.ones(data.shape[0]), data[:,:2], np.prod(data[:,:2], axis=1), data[:,:2]**2]
C_2,_,_,_ = scipy.linalg.lstsq(A_2, data[:,2])
    

Z_2 = np.dot(np.c_[np.ones(XX.shape), XX, YY, XX*YY, XX**2, YY**2], C_2).reshape(X.shape)
    
A = np.c_[np.ones(data.shape[0]), data[:,0], data[:,1], data[:,0]**2, np.prod(np.c_[data[:,0],data[:,1]],axis=1),\
    data[:,1]**2, data[:,0]**3, np.prod(np.c_[data[:,0]**2,data[:,1]],axis=1),\
    np.prod(np.c_[data[:,0],data[:,1]**2],axis=1), data[:,1]**3]

C,_,_,_ = scipy.linalg.lstsq(A, data[:,2]) # coefficients
    
Z = np.dot(np.c_[np.ones(XX.shape), XX, YY, XX**2, XX*YY, YY**2, XX**3, XX**2*YY, XX*YY**2, YY**3], C).reshape(X.shape)
    

AA = np.c_[np.ones(data.shape[0]), data[:,0], data[:,1], data[:,0]**2, np.prod(np.c_[data[:,0],data[:,1]],axis=1),\
    data[:,1]**2, data[:,0]**3, np.prod(np.c_[data[:,0]**2,data[:,1]],axis=1),\
    np.prod(np.c_[data[:,0],data[:,1]**2],axis=1), data[:,1]**3, data[:,0]**4,\
              np.prod(np.c_[data[:,0]**3,data[:,1]],axis=1), np.prod(np.c_[data[:,0],data[:,1]**3],axis=1),\
             data[:,1]**4, np.prod(np.c_[data[:,0]**2,data[:,1]**2],axis=1),\
              np.prod(np.c_[data[:,0]**3,data[:,1]**2],axis=1), np.prod(np.c_[data[:,0]**2,data[:,1]**3],axis=1)]

CC,_,_,_ = scipy.linalg.lstsq(AA, data[:,2])
    
ZZ = np.dot(np.c_[np.ones(XX.shape), XX, YY, XX**2, XX*YY, YY**2, XX**3, XX**2*YY, XX*YY**2, YY**3,\
                    XX**4, XX**3*YY, XX*YY**3, YY**4, XX**2*YY**2, XX**3*YY**2, XX**2*YY**3], CC).reshape(X.shape)
    
    
A_id = np.c_[np.ones(data.shape[0]), data[:,0], data[:,1], data[:,0]**2, np.prod(np.c_[data[:,0],data[:,1]],axis=1),\
    data[:,1]**2, data[:,0]**3, np.prod(np.c_[data[:,0]**2,data[:,1]],axis=1),\
    np.prod(np.c_[data[:,0],data[:,1]**2],axis=1), data[:,0]**4,\
              np.prod(np.c_[data[:,0]**3,data[:,1]],axis=1), np.prod(np.c_[data[:,0]**2,data[:,1]**2],axis=1)]
    
C_id,_,_,_ = scipy.linalg.lstsq(A_id, data[:,2])
    
Z_id = np.dot(np.c_[np.ones(XX.shape), XX, YY, XX**2, XX*YY, YY**2, XX**3, XX**2*YY, XX*YY**2,
                     XX**4, XX**3*YY, XX**2*YY**2], C_id).reshape(X.shape) # YY**4
```
